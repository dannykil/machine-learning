# Prompt Engineering

### 1. Be Concise(간결함)

```python
# 간결하지 못한 버전
p1 = """
친절하고 유머러스한 말투로 답변해줘. 
나는 개발자인데 초보자를 위해 설명하듯 말해줘. 존댓말 써주고, 가끔 농담도 섞어줘. 
전문용어는 쉽게 풀어줘. 긍정적인 분위기로 말하고 너무 딱딱하게는 하지마. 
질문에 필요한 정보는 꼭 다 알려줘. 궁금한 거 다 알려줘.
"""

# 간결한 버전
p2 = """
당신은 이제부터 친절하고 유머러스한 개발자 멘토입니다.
대상: 파이썬 초보 개발자
어조: 존댓말 사용, 긍정적이고 비격식적인 분위기, 가끔 가벼운 농담 허용.
목표: 전문 용어를 쉽게 풀어서 설명하며, 질문에 대한 모든 필수 정보를 빠짐없이 제공합니다.
"""
```

```python
p1res = model.generate_content(p1)
print(p1res.text)
```

```python
p2res = model.generate_content(p2)
print(p2res.text)
```

### 2. Be Specific and Well Defined

```python
p1 = """
재활용 중요해. 어떻게 잘 할 수 있을까? 방법 알려줘.
"""

p2 = """
저는 서울 강남구에 거주하는 주부입니다.
목표: 저희 아파트 단지 내 투명 페트병 분리수거율을 50% 이상 높일 수 있는 실질적인 방안을 찾고 있습니다.

요청 내용:
투명 페트병 올바른 분리수거 방법 (뚜껑, 라벨 분리 여부 등)을 단계별로 상세히 설명해주세요.
주민들의 참여를 유도할 수 있는 창의적인 아이디어를 3가지 이상 제안해주세요. (예: 인센티브, 교육 캠페인, 홍보 문구 등)
성공적인 해외/국내 아파트 단지의 투명 페트병 재활용 사례를 1~2개 포함해주세요.
답변은 명확하고 실용적이며, 주민들이 쉽게 이해하고 적용할 수 있는 언어로 작성해주세요.
"""
```

```python
p1res = model.generate_content(p1)
print(p1res.text)
```

```python
p2res = model.generate_content(p2)
print(p2res.text)
```

### 3. Ask One Question at a Time

```python
p1 = """
재활용 중요해. 어떻게 잘 할 수 있을까? 방법 알려줘. 20단어 내외로 설명해줘
"""

p2 = """
자연을 보호해야하는데 어떻게 해야할까?. 20단어 내외로 설명해줘
"""
```

```python
p1res = model.generate_content(p1)
print(p1res.text)

p2res = model.generate_content(p2)
print(p2res.text)
```

```python
generation_config = GenerationConfig(temperature=1.0)

p1 = """
오늘 날짜와 현재시간 알려줘
"""

p1res = model.generate_content(p1, generation_config=generation_config)
print(p1res.text)
```

### Shot Prompting(one shot, few shot, multi shot)
> 예시 제공 횟수 : 우리가 LLM으로 부터 원하는 콘텐츠 형태로 설정할 수 있음

```python
# 축구: 발로 공을 차서 골대에 넣는 팀 경기.
# 농구: 손으로 공을 던져 림에 넣는 박진감 넘치는 팀 스포츠.
# 야구: 투수가 던진 공을 타자가 쳐서 점수를 내는 인기 팀 게임.
# 수영: 물에서 팔다리를 이용해 유영하는 전신 운동.
# 테니스: 라켓으로 공을 주고받으며 네트를 넘기는 개인 또는 복식 경기.
    
p1 = """ 아래의 Snippet이 축구, 농구, 야구, 수영, 테니스 중 어떤 종목인지 알려줘.

예시 : 발로 공을 차서 골대에 넣는 팀 경기.
종목 : 축구

Snippet : 물에서 팔다리를 이용해 유영하는 전신 운동.
"""

p1res = model.generate_content(p1, generation_config=generation_config)
print(p1res.text)
```
