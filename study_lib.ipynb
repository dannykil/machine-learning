```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 11)
y1 = 2 * x + 1.5
y2 = y1 * rd.random() # 0.0 ~ 1.0 사이의 난수 생성
print("x : ", x)
print("y1 : ", y1)
print("y2 : ", y2) 
```

```python
x = np.linspace(0, 10, 11)
y1 = 2 * x + 1.5
num = len(x)
y2 = np.zeros(num)
print(y2)

rd.seed() # 난수 초기화(사용하지 않아도 무방)
```

```python
# 난수 생성
for i in range(0, num):
    noise = rd.uniform(-3, 3) # -3과 3 사이에서 난수 생성
    # noise = rd.uniform(0, 1)
    y2[i] = y1[i] + noise

# matplotlib을 사용한 그래프 그리기
plt.scatter(x, y1, c='black', s=80)
plt.scatter(x, y2, c='red', s=80) # noise
plt.show()

print(x)
print(y1)
print(y2)
```

```python
# 선형 회귀 모델
from sklearn.linear_model import LinearRegression

# 선형 모델 선택
model = LinearRegression()

# 모델 훈련
# x를 2차원 배열로 reshape
model.fit(x.reshape(-1, 1), y2)
# model.fit(x, y1)

# 키프로스에 대해 예측을 만듭니다
X_new = [[5]]

print(model.predict(X_new)) 
```

```python
# k-최근접 이웃 회귀
from sklearn.neighbors import KNeighborsRegressor 

# k-최근접 이웃 회귀 모델 선택
model = KNeighborsRegressor(n_neighbors=3)

# 모델 훈련
model.fit(x.reshape(-1, 1), y2)
# model.fit(x, y)

# 키프로스에 대해 예측을 만듭니다
X_new = [[37_655.2]]

print(model.predict(X_new)) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 여러 개의 서브플롯 생성
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# 각 서브플롯에 데이터 그리기
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Subplot 1')
axs[0, 0].set_xlabel('X-axis Label Long Long Long')
axs[0, 0].set_ylabel('Y-axis')

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Subplot 2 Very Long Title')

axs[1, 0].plot(x, y1 + y2, color='green')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

axs[1, 1].scatter(x, np.random.rand(100), color='purple')
axs[1, 1].set_title('Subplot 4')

# 그림 제목 추가
fig.suptitle('Figure Title Very Very Long')

# tight_layout() 호출
plt.tight_layout()

# 그림 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label='Sine Wave')
ax.set_xlabel('X')
ax.set_ylabel('Sine(X)')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1)) # 범례를 그림 바깥에 배치

plt.tight_layout(rect=[0, 0, 0.9, 1]) # 오른쪽 공간 확보

plt.show()
```
