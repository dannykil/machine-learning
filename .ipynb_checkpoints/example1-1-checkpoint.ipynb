```python
import matplotlib.pyplot as plt
import pandas as pd
# from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor # k-최근접 이웃 회귀
```

```python
# data_root = "https://github.com/ageron/data/raw/main/"
# https://github.com/ageron/data/tree/main/lifesat
# data_root = "https://github.com/ageron/data/tree/main/"
# data_root = "https://github.com/ageron/data/tree/main/lifesat/lifesat.csv"
# data_root = "https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat.csv"
data_root = "/Users/danniel.kil/Documents/workspace/machine-learning/lifesat.csv"
lifesat = pd.read_csv(data_root)
# lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
# lifesat = pd.read_csv("https://github.com/ageron/data/tree/main/lifesat/lifesat_full.csv")
x = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values
# print("x : ", x)
# print("y : ", y)
```

```python
# 데이터를 그래프로 나타냅니다.
lifesat.plot(kind='scatter', grid=True, x="GDP per capita (USD)", y="Life satisfaction")
# plt.axis([23_500, 62_500, 4, 9])
plt.axis([13_500, 72_500, 4, 9])
plt.show()
```

```python
# 선형 모델 선택
# model = LinearRegression()

# k-최근접 이웃 회귀 모델 선택
model = KNeighborsRegressor(n_neighbors=3)

# 모델 훈련
model.fit(x, y)

# 키프로스에 대해 예측을 만듭니다
X_new = [[37_655.2]] # 2020년 키프로스 1인당 GDP

# 출력 6.30165767 (LinearRegression)
# 출력 6.33333333 (KNeighborsRegressor)
print(model.predict(X_new)) 
```
